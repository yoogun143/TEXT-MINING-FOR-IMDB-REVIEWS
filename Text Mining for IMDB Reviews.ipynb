# Dataset description
IMDB, Internet Movie database is an online database of information related to word films, television programs, etc. with fan reviews and ratings. This site is extremely well-known among movie fanatics.

The dataset is the labeled dataset of 50,000 IMDB reviews, which has binary sentiment. It means that IMDB rating less than 5 results in a sentiment score of 0 and IMDB rating no less than 7 have a sentiment score of 0. This project will try to investigate the prevalent words in each comment by using word cloud. The data could also be examined further by distinguish the different and common words in each sentiment to see the words that are used frequently with each sentiment. This dataset also goes along with a test dataset, which is unlabeled. The data mining technique and the sentimental analysis could be used to create a logistic algorithm to predict the likelihood of negative or positive judgement of audiences.

Dataset: https://www.kaggle.com/ymanojkumar023/kumarmanoj-bag-of-words-meets-bags-of-popcorn

# Methodology
The process of text mining for this IMDB reviews includes:
-	Giving the overview of the prevalent words with bar graphs and word cloud.
-	Comparing positive and negative reviews based on common and different words between two sentiments using commonality cloud and comparison cloud respectively.
-	Giving a different insight into common words between two sentiments by using pyramid plot.
-	Finding cluster of words in sentiment 1 by using hierarchical clustering.
-	Plotting the association for a specific word in sentiment 1.

## Subset data

Since the data consists of 25000 observations, we need to subset the data to make is easier to investigate. Moreover, it is appropriate to maintain the proportion of negative and positive comments in the data. Therefore, we will extract 1000 from 12500 positives and 1000 from 12500 negatives before combining them into the final sample.

Firstly, we extract the index of sentiment 1 and 0 in the data by using “which” function, with the parameter is a logical vector. The result of this function is an integer vector with length equals to sum(x)



```r
# Get index of sentiment 1 and 0 reviews
index1 <- which(IMDBreviews$sentiment == 1)
index0 <- which(IMDBreviews$sentiment == 0)
```

Next, we set seed = 1 to make sure that the result of random selection is reproducible. The “sample” function will help us to acquire 1000 random indexes from overall indexes.

```r
# Get index of sentiment 1 and 0 samples
set.seed(1)
index1_sample <- sample(index1, 1000)
index0_sample <- sample(index0, 1000)
```

After that, we create a new character object which combines all sample strings by pasting each string separated by a space. Those strings are extracted from the data by the logical vector index above.

Finally, we paste those two character objects by using c() function.

```r
# Subset IMDBreviews with each index sample and then combine them to the final sample vector
IMDBreviews1 <- paste(IMDBreviews[index1_sample,"review"], collapse = " ")
IMDBreviews0 <- paste(IMDBreviews[index0_sample,"review"], collapse = " ")
IMDBsample <- c(IMDBreviews1, IMDBreviews0)
```

## Prepare frequency data for visualization

Initially, we have to create a Corpus object, which is the main core of text mining technique. The main structure for managing documents in tm is a so-called Corpus, representing a collection of text documents. A corpus is an abstract concept, and there can exist several implementations in parallel. 

The default implementation is the so-called VCorpus (short for Volatile Corpus) which realizes a semantics as known from most R objects: corpora are R objects held fully in memory. Within the corpus constructor, x must be a Source object which abstracts the input location.

```r
# Load the new vector and then make the VCorpus object
reviews_source <- VectorSource(IMDBsample)
reviews_corpus <- VCorpus(reviews_source)
```

Next, tm package is useful in pre-processing data. We will create a new custom function with one parameter “corpus”. Inside the function, we use tm_map to apply the defined cleaning technique to each element in corpus. 

The cleaning method in this paper is listed below:
-	*stripWhitespace*: Strip extra whitespace from a text document. Multiple whitespace characters are collapsed to a single blank.
-	*removePunctuation*: Remove punctuation marks from a text document.
-	*content_transformer(tolower)*: transform character from upper to lower to make sure there is no distinction between upper and lower character
-	*removewords*: remove words from text documents. The glossary used here is stopwords(“en”), which is prevalent words in English as well as user-defined words: “film”, “movie”, “one”. This function helps eliminate undesirable words in the result.
-	*Iconv(enc2utf8(x), sub = “byte)*: convert a character vector between encoding, it will convert elements of character vectors to the UTF-8. This function will help to address the problem of system cannot recognize uncommon symbols like Emoji, which is prevalent in social media.


```r
# Define clean_corpus function which will clean the text for analyzing
clean_corpus <- function(corpus){
  corpus <- tm_map(corpus, removePunctuation)
  corpus <- tm_map(corpus, content_transformer(tolower))
  corpus <- tm_map(corpus, removeWords, c(stopwords("en"), "film", "movie", "one"))
  corpus <- tm_map(corpus, content_transformer(function(x) iconv(enc2utf8(x), sub = "byte")))
  corpus <- tm_map(corpus, stripWhitespace)
  return(corpus)
}
```

The function is then applied to clean the corpus.

```r
# Clean VCorpus object with the clean-corpus function
clean_corp <- clean_corpus(reviews_corpus)
```

The next work is to make a TermDocumentMatrix from the corpus object. A document-term matrix or term-document matrix is a mathematical matrix that describes the frequency of terms that occur in a collection of documents. In a term-document matrix, rows correspond to terms in the collection and columns correspond to documents.

```r
# Make TermDocumentMatrix from the cleaned VCorpus above and assign 
#column names for the newly created TermDocumentMatrix
reviews_tdm <- TermDocumentMatrix(clean_corp)
colnames(reviews_tdm) <- c("Sentiment 1", "Sentiment 0")
```

The summary of the term-document matrix is represented below.

```r
inspect(reviews_tdm)
```

Following on from that, we transform the TermDocumentMatrix into a matrix and then count the number of each term by taking sum of rows of the matrix and then sort by decreasing order. 

```r
# Transform TermDocumentMatrix into a matrix
reviews_m <- as.matrix(reviews_tdm)

# Count the number of each term by taking sum of rows of the matrix and then sort by decreasing order
term_frequency <- rowSums(reviews_m)
term_frequency <- sort(term_frequency, decreasing = TRUE)
```

## Prepare frequency data for wordcloud

To begin with, we transform numeric object “term-frequency” into a dataframe because wordcloud only accepts dataframe object as an input.

Next, we will make the wordcloud more appealing by customizing the color ramp. We can display all possible brewer by using display.brewer.all()

```r
# Display all possible brewer and then create a new "yellow_red" color ramp
display.brewer.all()
yellow_red <- brewer.pal(9, "YlOrRd")
yellow_red <- yellow_red[-(1:2)]
```

There are 3 types of palettes, sequential, diverging, and qualitative.
- *Sequential palettes* are suited to ordered data that progress from low to high. 
- *Diverging palettes* put equal emphasis on mid-range critical values and extremes at both ends of the data range. 
- *Qualitative palettes* do not imply magnitude differences between legend classes, and hues are used to create the primary visual differences between classes. Qualitative schemes are best suited to representing nominal or categorical data.


In this case, we will choose a **sequential palette** to suit our purpose of sequential order. The new palette is chosen and assigned to “yellow_red” variable.

## Preparation for pyramid plot

Commonality.cloud() could be misleading as words can be represented disproportionately even if they are shared. **Pyramid plot** can overcome this limitation by visualizing the number of words in each document for each common term. That is to say, pyramid plot is a worthwhile upgrade for the commonality cloud.

First, we subset the reviews_m matrix by extracting those terms with more than 0 words in each document. It can be done by using a logical expression in second argument of subset() function. The subset will then be assigned to common_words

```r
# Subset the matrix with the condition that the term appears in both sentiments
common_words <- subset(reviews_m, reviews_m[, 1] > 0 & reviews_m[, 2] > 0)
```

After that, we calculate the absolute difference of each term by using abs() function for the minus of column 1 and column 2 of common_words matrix. We then bind this difference to current matrix by using cbind() function and sort the matrix by difference column.

```r
# Calculating the difference of each term in two sentiments, assgin the result into a new column in the matrix 
#and sort the matrix by the difference column
difference <- abs(common_words[,1] - common_words[,2])
common_words <- cbind(common_words, difference)
common_words <- common_words[order(common_words[, 3], decreasing = TRUE), ]
```

The final step is to create a dataframe called top25_df which consists of 25 terms along with the number of words in each sentiment.

```r
# Create top25_df dataframe which consists of 25 terms associating with the number of words in each sentiment
top25_df <- data.frame(x = common_words[1:25, 1], 
                       y = common_words[1:25, 2], 
                       labels = rownames(common_words[1:25, ]))
```

## Plot pyramid plot

We plot pyramid plot by using pyramid.plot() function from plotrix package. The first 3 arguments are the elements of top25_df, the gap argument defines the gap between bars

```r
# Create pyramid plot of top25 common words
pyramid.plot(top25_df$x, top25_df$y,
             labels = top25_df$labels, gap = 150,
             top.labels = c("Sentiment 1", "Words", "Sentiment 0"),
             main = "Words in Common", laxlab = NULL,
             raxlab = NULL, unit = NULL)
```

It can be inferred from the plot that there may be a clear distinction in some common words between two sentiments. For instance, the word “bad” can be common but it is results mostly from negative sentiment. The same pattern happens with “even” word and “just” word. The word “great”, “love”, “also”, “best” is the main driving force to make reviews become commonplace in positive reviews and in total.

## Create Term Document Matrix for Sentiment 1

Initially, IMDB reviews with Sentiment 1 will be extracted from the original data with index1_sample created in step 3. After that, the term document matrix is created from “IMDBrevies1_ “vector under the same procedure explained in step 4.

```r
# Create Term Document Matrix for Seniment 1
IMDBreviews1_ <- IMDBreviews[index1_sample, "review"]$review
reviews1_source <- VectorSource(IMDBreviews1_)
reviews1_corpus <- VCorpus(reviews1_source)
clean1_corp <- clean_corpus(reviews1_corpus)
reviews1_tdm <- TermDocumentMatrix(clean1_corp)
```

## Reduce the sparse in the Term Document Matrix and create data frame for clustering

There are two reasons why the Term Document Matrix needs to get rid of sparse terms. 

First, sparse terms are the terms contain mostly zeros in all documents, meaning that it may not bring a significant added value for analysis. 

Secondly, the original Term Document Matrix contains a gigantic number of Terms and it will be hard to conclude some insights from an over-cluttered object.

The removeSparseTerms() function works with a Term Document Matrix with the second “parse” argument determines the maximum percentage of empty elements for a term.

```r
# Remove Number of words in Term Document Matrix
reviews1_tdm2 <- removeSparseTerms(reviews1_tdm, sparse = 0.8)

# Create dataframe from the Term Document Matrix
reviews1_m <- as.matrix(reviews1_tdm2)
reviews1_df <- as.data.frame(reviews1_m)
```

It is recognized that the number of terms has declined from 21159 terms to 32 terms for examination and the sparsity reduces from 100% to 73%.

## Apply hierarchical clustering method

Firstly, the **distance matrix** is calculated, which is the difference between each row of the Term Document Matrix. The default method used here is Euclidean distance, the usual distance between two vectors

```r
# Calculate distance matrix
reviews1_dist <- dist(reviews1_df)
```

Following on from that, using hclust() function, the hierarchical clustering object will be created and assigned to “hc” object. This function takes the distance matrix calculated above as the first argument and the method as the second one. 

There are 4 main methods or linkage criteria to create a hierarchical clustering which are: 
-	Complete linkage: the distance between two clusters is defined as the longest distance between two points in each cluster
-	Single linkage: the distance between two clusters is defined as the shortest distance between two points in each cluster.
-	Mean or average linkage: the distance between two clusters is defined as the average distance between each point in one cluster to every point in the other cluster
-	Centroid linkage: the distance between two clusters is defined as the distance between two cluster centroids


```r
# Create hc
hc <- hclust(reviews1_dist, method = "complete")
```

## Plot the dendrogram

The dendrogram from hierarchical clustering object can be easily plotted with plot() function.

To add more features to the current dendrogram and make the clusters more distinct, the rect.dendrogram() function is used. The first argument is the dendrogram object transformed from hc. “k” defines the number of clusters desired, which is 5 in this case. The last two arguments determine the boundary for those clusters 


```r
# Plot the dendrogram
plot(hc)
rect.dendrogram(as.dendrogram(hc), k = 5, border = "red", density = 5)
```

With 5 clusters, the dendrogram is cut at the height of 38. The terms in the same cluster signifies that those often appears together. The difference in height is the distance between two terms calculated through Euclidean distance. In other words, the smaller the heights, the closest in the frequency of concurrent appearance.

For example, in cluster 5, “time” is in the same clutter as “love”. There is a high possibility that for positive reviews of “love” story, the romantic and precious time could be mentioned and vice versa. In the fourth cluster, “people” goes along with “life” and “see”. It may be inferred that those movies revolving around “life” could involves the social network around “people” when they “see” each other.

Otherwise, “story” is in its own cluster, meaning that the term “story” doesn’t often appear with “love”, or “people” at the same time. Another explanation could be “story” appears in almost any documents so there is negligible connection with other terms.


## Find correlation for a specific word

If word clustering examines the relationship among a set of terms, findAssocs() function could be helpful in determining the *correlation for a specific word*. This function has 3 arguments. The first one is the Term Document Matrix, the second one is the term need to find association, the last one is the lower correlation limits of each term from zero to one. It means that those terms with correlation with “character” more than 0.25 will be included in the result.

```r
# Create associations
associations <- findAssocs(reviews1_tdm, "character", 0.25)
```

Next, the associations list will be transformed into a data frame by using list_vect2df() function in qdap package.

```r
# Convert above list to dataframe
associations_df <- list_vect2df(associations)[,2:3]
```

## Visualize association

Using ggplot(), we can visualize the data, there are 3 layers: the base layer, the geom_point() layer and the theme layer to define the scatter plot.

```r
# Plot the associations_df values
ggplot(associations_df, aes(y = associations_df[, 1])) + 
  geom_point(aes(x = associations_df[, 2]), 
             data = associations_df, size = 3) + 
  theme_bw()
```

As can be seen, the word “character” is most related to the word “incomplete” with the correlation of 0.29, followed by “played” and “character”. It could be inferred that for positive reviews, those “character” could be “incomplete”, “impulsive” or in “development”
